```python
import pandas as pd
import numpy as np
import swifter
```

```python
df = pd.read_parquet('data/dataset_merged_cleaned.parquet')
df.rename(columns={'stock_tickers': 'ticker'}, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
gics_sector_etfs = {
  "Energy": "VDE",  # Vanguard Energy ETF
  "Materials": "VAW",  # Vanguard Materials ETF
  "Industrials": "VIS",  # Vanguard Industrials ETF
  "Consumer Discretionary": "VCR",  # Vanguard Consumer Discretionary ETF
  "Consumer Staples": "VDC",  # Vanguard Consumer Staples ETF
  "Health Care": "VHT",  # Vanguard Health Care ETF
  "Financials": "VFH",  # Vanguard Financials ETF
  "Information Technology": "VGT",  # Vanguard Information Technology ETF
  "Communication Services": "VOX",  # Vanguard Communication Services ETF
  "Utilities": "VPU",  # Vanguard Utilities ETF
  "Real Estate": "VNQ"  # Vanguard Real Estate ETF
}

# Function to append sector ETF tickers to the ticker list
def append_sector_etf(row):
  sectors_list = row['sectors']
  tickers_list = row['ticker']
  
  updated_tickers = list(tickers_list) if isinstance(tickers_list, (list, np.ndarray)) else []

  if isinstance(sectors_list, (list, np.ndarray)):
    for sector in sectors_list:
      if sector in gics_sector_etfs:
        etf_ticker = gics_sector_etfs[sector]
        if etf_ticker not in updated_tickers:
          updated_tickers.append(etf_ticker)
      
  return updated_tickers

df['ticker'] = df.swifter.apply(append_sector_etf, axis=1)
```

```python
print(df.shape)
df.head()
```

```python
# Explode the 'ticker' column to handle multiple 'ticker' per article
df_exploded = df.explode('ticker')

# Drop rows where 'sectors' might be NaN after exploding (if the original list was empty)
df_exploded = df_exploded.dropna(subset=['ticker'])

print(df_exploded.shape)
df_exploded.head()
```

```python
# group data by hour and aggregate the columns

aggregation_dict = {
  'time_published': 'count',  # for counting the number of rows

  ### sentiment columns ###
  'title_sentiment_finbert': 'mean',
  'summary_sentiment_finbert': 'mean',
  'title_sentiment_vader': 'mean',
  'summary_sentiment_vader': 'mean',

  ### topic columns ###
  'topic_announcements': 'sum',
  'topic_market_movements': 'sum',
  'topic_legal_actions': 'sum',
  'topic_investment_sentiment': 'sum',
  'topic_earnings': 'sum',
  'topic_analyst_ratings': 'sum',
  'topic_macro_econ': 'sum',
  'topic_research_insights': 'sum',

  ### NER columns ###
  'NER_US': 'sum',
  'NER_China': 'sum',
  'NER_India': 'sum',
  'NER_Russia': 'sum',
  'NER_Canada': 'sum',
  'NER_UK': 'sum',
  'NER_Trump': 'sum',
  'NER_ElonMusk': 'sum',
  'NER_JoeBiden': 'sum',
  'NER_FED': 'sum',
  'NER_Ukraine': 'sum',
  'NER_Japan': 'sum',
}


# Group by hour (using time_published) and sectors
df_grouped = df_exploded.groupby([pd.Grouper(key='time_published', freq='h'), 'ticker']).agg(aggregation_dict)
df_grouped.rename(columns={'time_published': 'news_count'}, inplace=True)
df_grouped.reset_index(inplace=True)
df_grouped.rename(columns={'time_published': 'timestamp'}, inplace=True)
```

```python
print(df_grouped.shape)
df_grouped.head()
```

```python
tickers_list = df_grouped['ticker'].unique()
tickers_list
```

```python
# Retrieve return of ticker at that timestamp

df_return = pd.read_parquet('../data_collection/data_warehouse/stock_return.parquet')
df_return.reset_index(inplace=True)
df_return = df_return[df_return['ticker'].isin(tickers_list)]
df_return = df_return[df_return['timestamp'] >= '2022-03-01']

df = pd.merge(df_grouped, df_return[['timestamp', 'ticker', 'return_forwarded']],
           on=['timestamp', 'ticker'],
           how='right')

print(df.shape)
df.head()
```

```python
df.to_parquet('data/dataset_features_return.parquet', index=False)
```
