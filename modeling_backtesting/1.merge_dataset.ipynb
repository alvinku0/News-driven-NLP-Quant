```python
import pandas as pd
```

```python
df_news = pd.read_parquet('../data_collection/data_warehouse/news_articles.parquet')
df_sentiment = pd.read_parquet('../features_extraction/data/news_sentiment.parquet')
df_NER = pd.read_parquet('../features_extraction/data/NER.parquet')
df_preprocess = pd.read_parquet('../features_extraction/data/text_preprocessed.parquet')
df_topic = pd.read_parquet('../features_extraction/data/topic_LDA.parquet')
df_tickers = pd.read_parquet('../features_extraction/ticker/ticker_result.parquet')
df_sectors = pd.read_parquet('../features_extraction/data/sectors.parquet')
```

```python
df_news = df_news[['title', 'summary', 'time_published', 'source']]

# merge all dataframes
df = df_preprocess.merge(df_news, left_index=True, right_index=True)
df = df.merge(df_sentiment, left_index=True, right_index=True)
df = df.merge(df_NER, left_index=True, right_index=True)
df = df.merge(df_topic, left_index=True, right_index=True)
df = df.merge(df_tickers, left_index=True, right_index=True)
df = df.merge(df_sectors, left_index=True, right_index=True)

df.index.name = 'newsID'

print(df.shape)
df.head()
```

```python
df.drop(columns=['title_preprocessed', 'summary_preprocessed'], inplace=True)
```

```python
df.to_parquet('data/dataset_merged.parquet')
```
